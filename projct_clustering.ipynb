```python
# Function to check if conversion to datetime is possible for "FFP_DATE", "FIRST_FLIGHT_DATE", and "LAST_FLIGHT_DATE
def check_date_format(column):
    try:
        pd.to_datetime(column, errors='raise')
        return True
    except (ValueError, TypeError):
        return False
```

```python
# Check each date column for invalid formats
date_columns = ["FFP_DATE", "FIRST_FLIGHT_DATE", "LAST_FLIGHT_DATE"]
errors = {col: check_date_format(df[col]) for col in date_columns}
errors
```

```python
# Print results and identify invalid date formats
invalid_rows = []
for col, is_valid in errors.items():
    if not is_valid:
        print(f"Column '{col}' contains invalid date formats.")
        invalid_rows.extend(df[~df[col].apply(lambda x: check_date_format(pd.Series([x])))].index)
```

```python
# Drop rows with invalid date formats
if invalid_rows:
    df = df.drop(index=invalid_rows).reset_index(drop=True)
    print(f"Dropped {len(invalid_rows)} rows with invalid date formats.")
else:
    df_cleaned = df.copy()
    print("No invalid date formats found.")
```

```python
df_clean = df.copy() 

# Impute rows with mode & mean
df_clean['WORK_CITY'].fillna(df_clean['WORK_CITY'].mode()[0], inplace=True)
df_clean['WORK_PROVINCE'].fillna(df_clean['WORK_PROVINCE'].mode()[0], inplace=True)
df_clean['WORK_COUNTRY'].fillna(df_clean['WORK_COUNTRY'].mode()[0], inplace=True)
df_clean['GENDER'].fillna(df_clean['GENDER'].mode()[0], inplace=True)

df_clean['AGE'].fillna(df_clean['AGE'].mean(), inplace=True)
df_clean['SUM_YR_1'].fillna(df_clean['SUM_YR_1'].mean(), inplace=True)
df_clean['SUM_YR_2'].fillna(df_clean['SUM_YR_2'].mean(), inplace=True)

# Check missing values
df_clean.isna().sum()
```

```python
# Change date columns from object to datetime
df_clean['FFP_DATE'] = pd.to_datetime(df_clean['FFP_DATE'])
df_clean['LOAD_TIME'] = pd.to_datetime(df_clean['LOAD_TIME'])
```

```python
# Extract loyalty / membership duration from FFP_DATE - LOAD_TIME (in months)
df_clean['LOYALTY'] = (df_clean['LOAD_TIME'].dt.year - df_clean['FFP_DATE'].dt.year) * 12 + (df_clean['LOAD_TIME'].dt.month - df_clean['FFP_DATE'].dt.month)

# Display the first few rows to check
print(df_clean[['FFP_DATE', 'LOAD_TIME', 'LOYALTY']].sample(5))
```

The main goal of classifying customers based on their characteristics is to understand and manage customer interactions across key dimensions important to the business. Based on this foundation, the following features have been selected for clustering:

The "LOYALTY" feature represents Length: The duration of a customer's membership, from the time of first registration to the observation period (the longer the membership, the more "loyal" or valuable the customer is considered).

The "AVG_INTERVAL" feature represents Interval: The time between flights (a smaller interval means the customer has flown recently, which is preferable as smaller intervals are better).

The "FLIGHT_COUNT" feature represents Frequency: The number of flights taken by the customer during the observation period (the higher the count, the better, indicating frequent flyers).

The "SEG_KM_SUM" feature represents Distance Traveled: The total distance flown during the observation period (the greater the distance, the higher the cost paid by the customer).

```python
iqr_columns = ['FLIGHT_COUNT', 'SEG_KM_SUM', 'AVG_INTERVAL']
       
def remove_outliers_iqr(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

# Remove outliers from the specified columns using the IQR method and update df_clean
df_clean = remove_outliers_iqr(df_clean, iqr_columns)

# Display the shape of the dataframe after removing outliers
print("Shape of dataframe after removing outliers:", df_clean.shape)
```

```python
X_columns = ['LOYALTY', 'FLIGHT_COUNT', 'SEG_KM_SUM', 'AVG_INTERVAL']

X = df_clean[X_columns].values

from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(X)
new_df = pd. DataFrame (data = X_std, columns = X_columns)
new_df.describe ()
```

```python
from sklearn.cluster import KMeans
inertia = []

for i in range(1, 11):
    kmeans = KMeans (n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(new_df.values)
    inertia.append(kmeans.inertia_)

    
plt.figure(figsize=(10, 4))
plt.plot (inertia)
plt.grid()
```

```python
fig, ax = plt.subplots(figsize=(12,4))
sns.lineplot(x=range(1, 11), y=inertia, color="skyblue", linewidth = 3)
sns.scatterplot(x=range(1, 11), y=inertia, s=100, color='red')
plt.title("Inertia Score Elbow Method Kmean Clustering")
plt.grid()

plt.show()
```

From the Elbow graph above, it is observed that after the 4th point (k=5), the reduction in inertia becomes less significant. To confirm this further, we will examine the reduction in inertia at each step as the number of clusters increases.

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=42)
kmeans.fit(new_df.values)
# Add cluster labels to the dataframe
new_df['Cluster'] = kmeans.labels_
```

```python
new_df.sample(10)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3, random_state=42)

pca.fit(new_df)
pcs = pca.transform(new_df)

data_pca = pd.DataFrame(data = pcs, columns = ['PC 1', 'PC 2', 'PC 3'])
data_pca['ClusterPCA'] = new_df['Cluster']
data_pca.sample(5)
```

```python
fig, ax = plt.subplots(figsize=(12,5))

sns.scatterplot( x="PC 1", y="PC 2", hue="ClusterPCA", data=data_pca, edgecolor='black', linestyle='--',
                palette='viridis', ax=ax, s = 100)
plt.title("Segmentation with 5 clusters")
plt.show()
```

# Business Insight and Recomendations:

### Explanation of Clusters Based on Customer Characteristics:


**Cluster 0 (Inexperienced Flyers):**

Customers in this cluster are new members of the airline program.
They havenâ€™t taken many flights yet, although their average interval between flights is short.
These customers may include individuals who are just starting to explore or use air travel services, such as students or people beginning their professional careers.

**Cluster 1 (Rare Flyers):**

Customers in this cluster have been members longer than those in Cluster 0, but they still take relatively few flights and cover little distance.
Their average interval between flights is the highest among all clusters, indicating that they rarely fly.
Customers in this cluster might be individuals who only travel for personal reasons or take long vacations and rarely fly, or those who prefer other modes of transportation.

**Cluster 2 (Infrequent Flyers):**

Customers in this cluster have been members the longest compared to other clusters.
Despite the long membership duration, they still take a relatively small number of flights and cover limited distances.
Their average flight interval is moderate, suggesting they travel sporadically or only on special occasions.
Customers in this cluster might include individuals who prefer to travel only during special times, such as annual vacations or rare business trips.

**Cluster 3 (Regular Flyers):**

Customers in this cluster have recently become members but take more flights than average.
The total distance they travel is also significant.
Their average interval between flights is moderate, indicating regular travel, possibly for business or work purposes.
Customers in this cluster might include businesspeople or professionals who travel routinely for meetings or work-related purposes.

**Cluster 4 (Frequent Flyers):**

Customers in this cluster have the second-longest membership duration after Cluster 2.
They take the most flights and cover the highest total distance compared to other clusters.
Their average interval between flights is short, indicating they fly frequently.
These customers may include business executives who frequently travel internationally or domestically, or travelers who often fly for leisure or recreational activities.

### Business Recommendations for Each Cluster:

**Cluster 0 (Inexperienced Flyers):**

**Business Recommendation:** Offer special promotions to attract new customers, such as discounts on their first flight or bonus points upon registration. Focus on providing easily understandable services and avoid complexity, as they are new to air travel.

**Cluster 1 (Rare Flyers):**

**Business Recommendation:** Introduce long-term holiday packages or special promotions for rarely visited destinations. Provide flexible booking options and exclusive offers to encourage them to travel more frequently.

**Cluster 2 (Infrequent Flyers):**

**Business Recommendation:** Develop an engaging loyalty program with points or rewards that can be accumulated from each flight. Offer holiday packages or flexible policies that allow them to plan long trips or vacations effectively.

**Cluster 3 (Regular Flyers):**

**Business Recommendation:** Focus on premium services and fast business perks, such as priority access, exclusive airport lounges, or superior customer service. Offer loyalty programs that provide added value and convenience for frequent travelers.

**Cluster 4 (Frequent Flyers):**

**Business Recommendation:** Enhance exclusive services for frequent flyers, such as automatic upgrades, global VIP lounge access, or exclusive opportunities for luxurious travel experiences. Provide flexible policies that accommodate their frequent travel needs.
